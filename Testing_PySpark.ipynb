## Apache Arrow
Spark와 pandas 사이의 데이터를 효율적으로 공유하도록 하는 설정
* `sequence`: 모든 데이터를 순회해서 0부터 번호를 매기기 때문에, 큰 데이터셋에서 느리고 비효율적
* `distributed`: 인덱스를 Spark의 파티션 기반으로 분산하여 생성 -> 성능과 확장성 Good

```python
from pyspark.sql import SparkSession
import pyspark.pandas as ps


ps.set_option('compute.default_index_type', 'distributed')

spark = SparkSession.builder\
                    .config("spark.sql.execution.arrow.pyspark.enabled", "true") \
                    .getOrCreate()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder\
                    .config('spark.sql.execution.arrow.pyspark.enabled', True)\
                    .appName('Testing PySpark Example')\
                    .getOrCreate()
```

```python
## python dictionary
sample_data = [{"name": "John    D.", "age": 30},
  {"name": "Alice   G.", "age": 25},
  {"name": "Bob  T.", "age": 35},
  {"name": "Eve   A.", "age": 28}]


df = spark.createDataFrame(sample_data)
df.show()
```

```python
from pyspark.sql.functions import col, regexp_replace

# Remove additional spaces in name
# 특정 column에 대해 정규표현식이 적용된 df_transformed 데이터프레임을 반환
def remove_extra_spaces(df, column_name):
  df_transformed = df.withColumn(column_name, regexp_replace(col(column_name), "\\s+",""))

  return df_transformed

transformed_df = remove_extra_spaces(df,'name')

transformed_df.show()
```

# PySpark Application
Spark는 결과를 한 눈에 확인할 수 있지만 대규모 데이터에서는 비효율적
---
PySpark의 동작은 lazy evaluation이기 때문에, 명확한 단위 테스트(Unit Test)가 필요하다.
* uniitest
* pytest

* `assertDataFrameEqual()`: DataFrame의 schema와 값 비교까지 모두 수행한다.

```python
import pyspark.testing
## 'np.NaN' was removed in the NumPy 2.0 release. Use `np.nan` instead.
## NumPy 2.0 이상부터 np.NaN이 아닌 np.nan으로 처리하므로 문제 발생
## !pip install numpy==1.26.4
from pyspark.testing.utils import assertDataFrameEqual
import warnings
warnings.filterwarnings('ignore')

df1 = spark.createDataFrame(data = [('1', 1000), ('2', 3000)], schema = ['id', 'amount'])
df2 = spark.createDataFrame(data=[("1", 1000), ("2", 3000)], schema=["id", "amount"])

## DataFrame의 스키마 + 값 비교까지 모두 수행
assertDataFrameEqual(df1, df2)
```

```python
# Example 2
df1 = spark.createDataFrame(data=[("1", 0.1), ("2", 3.23)], schema=["id", "amount"])
df2 = spark.createDataFrame(data=[("1", 0.109), ("2", 3.23)], schema=["id", "amount"])

## rtol: 상대 오차를 허용하는 인자로, 여기서는 10% 허용한다는 뜻
assertDataFrameEqual(df1, df2, rtol=1e-1)  # pass, DataFrames are approx equal by rtol
```

```python
from pyspark.testing.utils import assertSchemaEqual
from pyspark.sql.types import StructType, StructField, ArrayType, DoubleType

# 스키마 구조, 필드 이름, 데이터타입, null 허용 등을 모두 비교
## ETL에서 입력 스키마와 출력 스키마가 동일해야하는 경우 assertSchemaEqual 활용 가
s1 = StructType([StructField("names", ArrayType(DoubleType(), True), True)])
s2 = StructType([StructField("names", ArrayType(DoubleType(), True), True)])
assertSchemaEqual(s1, s2)  # pass, schemas are identical
```

# unittest
* 코드가 많아지면 `assertSchemaEqual`하나로 테스트 하기 어려워짐
---
* Python 내장 라이브러리
* 테스트 케이스를 클래스로 깔끔하게 정리
* `@classmethod`를 이용해 테스트 클래스 전체에서 공유되는 SparkSession을 한번만 생성

```python
from pyspark.testing.utils import assertDataFrameEqual
from pyspark.sql.functions import regexp_replace, col
from pyspark.sql import SparkSession
import unittest # 파이썬 표준 테스트 프레임워크

# SparkSession을 관리하는 역할
class PySparkTestCase(unittest.TestCase):

  # @classmethod: 클래스 자체인 cls를 인자로 받음. 클래스 상태나 속성에 접근 가능
  # 데코레이터로 클래스에서 직접 호출 가능할 수 있게 하는 역할
  @classmethod
  def setUpClass(cls):
    cls.spark = SparkSession.builder.appName('Testing PySpark ETL').getOrCreate()

  # 테스트가 모두 끝난 후에 SparkSession을 종료하여 리소스
  @classmethod
  def tearDownClass(cls):
    cls.spark.stop()


def remove_extra_spaces(df, column_name):
  df_transformed = df.withColumn(column_name, regexp_replace(col(column_name), "\\s+"," "))

  return df_transformed

class TestTranformation(PySparkTestCase):
  def test_single_space(self):
      sample_data = [{"name": "John    D.", "age": 30},
                        {"name": "Alice   G.", "age": 25},
                        {"name": "Bob  T.", "age": 35},
                        {"name": "Eve   A.", "age": 28}]
      original_df = spark.createDataFrame(sample_data)
      transformed_df = remove_extra_spaces(original_df, 'name')

      expected_data = [{"name": "John D.", "age": 30},
        {"name": "Alice G.", "age": 25},
        {"name": "Bob T.", "age": 35},
        {"name": "Eve A.", "age": 28}]

      expected_df = spark.createDataFrame(expected_data)

      assertDataFrameEqual(transformed_df, expected_df)
```

```python
## unittest 테스트 프레임워크에서 테스트 실행 함수
if __name__ == '__main__':
    unittest.main(argv=[''], verbosity=0, exit=False)
```
